```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('train.csv')
df.replace("#REF!", pd.NA, inplace=True)
# Basic info
df.info()
df.describe(include='all')  # includes categorical too
```

```python
# Fill missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
# Convert Fare to numeric, coerce errors to NaN
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')

# Now safely fill missing values with mean
df['Fare'].fillna(df['Fare'].mean(), inplace=True)


# Create a new feature from Cabin (Deck)
df['Deck'] = df['Cabin'].astype(str).str[0]
df['Deck'] = df['Deck'].replace('n', 'Unknown')
```

```python
df['Survived'] = df['Survived'].fillna(0)  # or some appropriate value
df['Age'] = df['Age'].fillna(df['Age'].median())

# Plotting histograms for numeric columns
df[['Age', 'Fare']].hist(bins=20, figsize=(10, 4))
plt.tight_layout()
plt.show()






```

```python
# Count plots for categorical features
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Class')
plt.show()
```

```python

sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Sex')
plt.show()
```

```python
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title('Survival by Embarked Port')
plt.show()
```

```python
# Ensure 'Survived' and 'Age' are numeric
df['Survived'] = pd.to_numeric(df['Survived'], errors='coerce')
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Age vs Survival
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution by Survival')
plt.show()
```

```python
# Fare vs Survival
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare Distribution by Survival')
plt.show()
```

```python




# Pairplot of numerical columns
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived')
plt.show()


```

```python
# 1. Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
# 2. Survival by Passenger Class
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Class')
plt.show()
```

```python
# 4. Age Distribution by Survival
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age vs Survival')
plt.show()
```

```python
# 5. Correlation Heatmap
cols = ['Survived', 'Pclass', 'Age', 'Fare', 'SibSp', 'Parch']
for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert to numeric, coerce errors to NaN

corr = df[cols].dropna().corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

```
